```python
df = pd.read_excel("data.xlsx", skiprows=3)
df.tail()
```

```python
df.columns
```

```python
df.rename(columns={df.columns[0]: 'Region'}, inplace=True)
```

```python
df.columns
```

```python
# Оставим только колонку с регионами и годы
df = df.loc[:, ['Items'] + list(range(2011, 2025))]
```

```python
df
```

```python
df.rename(columns={'Items': 'Region'}, inplace=True)
```

```python
df_long = df.melt(id_vars='Region', var_name='Year', value_name='NetMigration')
df_long['Year'] = df_long['Year'].astype(int)
df_long.head()
```

```python
df_filtered = df_long[df_long['Region'] != 'Kyrgyz Republic']
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("Set2")

sns.lineplot(data=df_filtered, x='Year', y='NetMigration', hue='Region')

plt.title('Net Migration Trends by Region (2011–2024)\n(excluding Kyrgyz Republic)')
plt.xlabel('Year')
plt.ylabel('Net Migration')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

```python
# Выбор года для анализа
year = 2022

# Фильтрация по году
df_year = df_long[df_long['Year'] == year]

# Построение графика
plt.figure(figsize=(10, 6))
sns.barplot(data=df_year, x='NetMigration', y='Region', palette='coolwarm')
plt.title(f'Net Migration by Region in {year}')
plt.xlabel('Net Migration')
plt.ylabel('Region')
plt.tight_layout()
plt.show()

```

```python
heatmap_data = df_long.pivot(index='Region', columns='Year', values='NetMigration')
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', center=0, annot=True, fmt=".0f", linewidths=0.5)

plt.title('Net Migration Heatmap by Region and Year')
plt.xlabel('Year')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```
