```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Чтение Excel-файла (.xls) с указанием правильного движка
df = pd.read_excel("data.xlsx")

# Проверим первые строки
df.head()
```

```python
df = pd.read_excel('data.xlsx',skiprows=3)  # пропускаем строки до нужной
df.head()
```

```python
df = df[~df['Items'].isin(['Total arrivals', 'Of living in CIS countries'])]
df.head()
```

```python
df.rename(columns={'Items': 'Nation'}, inplace=True)
```

```python
df
```

```python
df = df.loc[:, ['Nation'] + list(range(2011, 2025))]
df
```

```python
df_long = df.melt(id_vars='Nation', var_name='Year', value_name='Arrivals')
df_long['Year'] = df_long['Year'].astype(int)
df_long.dropna(inplace=True)
df
```

```python
df_long['Arrivals'].replace("-", pd.NA, inplace=True)
```

```python
df_long['Arrivals'] = pd.to_numeric(df_long['Arrivals'], errors='coerce')
df_long.dropna(subset=['Arrivals'], inplace=True)
df_long
```

```python
# Увеличиваем стиль и читаемость
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("Set2")

# Строим график
sns.lineplot(data=df_long, x='Year', y='Arrivals', hue='Nation', marker='o')

# Оформление
plt.title('Migration to Kyrgyzstan by Nation of Origin (2011–2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Nation')
plt.tight_layout()
plt.show()
```

```python
df_area = df_long.pivot(index='Year', columns='Nation', values='Arrivals')
df_area.fillna(0, inplace=True)

df_area.plot.area(figsize=(14, 7), cmap='tab20')
plt.title('Migration to Kyrgyzstan by Nation (Stacked Area, 2011–2024)')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
heatmap_data = df_long.pivot(index='Nation', columns='Year', values='Arrivals')

sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False, linewidths=0.5)
plt.title('Migration to Kyrgyzstan by Nation and Year (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Pivot для area plot
df_area = df_long_final.pivot(index='Year', columns='Nation', values='Arrivals')
df_area.fillna(0, inplace=True)  # на случай пропусков

# Строим график
df_area.plot.area(figsize=(14, 7), cmap='tab20')
plt.title('Migration to Kyrgyzstan by Nation (Stacked Area, 2011–2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Arrivals')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Nation')
plt.tight_layout()
plt.show()
```

```python
year = 2023
top = df_long_final[df_long_final['Year'] == year].sort_values('Arrivals', ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(data=top, x='Arrivals', y='Nation', palette='viridis')
plt.title(f'Top 5 Nations Sending Migrants to Kyrgyzstan in {year}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_long_final, x='Nation', y='Arrivals')
plt.xticks(rotation=45)
plt.title('Distribution of Annual Arrivals by Nation (2011–2024)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df_long_final, x='Nation', y='Arrivals', jitter=True, size=6, palette='Set1')
plt.xticks(rotation=45)
plt.title('Yearly Migration Points by Nation')
plt.tight_layout()
plt.show()
```

```python
df_cum = df_long_final.pivot(index='Year', columns='Nation', values='Arrivals').fillna(0).cumsum()

df_cum.plot(figsize=(14, 6))
plt.title('Cumulative Migration to Kyrgyzstan by Nation (2011–2024)')
plt.xlabel('Year')
plt.ylabel('Total Arrivals')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
year = 2023
pie_data = df_long_final[df_long_final['Year'] == year].groupby('Nation')['Arrivals'].sum()
pie_data = pie_data[pie_data > 0]  # убрать нули

plt.figure(figsize=(8, 8))
plt.pie(pie_data, labels=None, startangle=140)
plt.title(f'Migration Share by Nation in {year}')
plt.legend(pie_data.index, loc='upper right')
plt.tight_layout()
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_long_final, x='Nation', y='Arrivals', inner='quartile', palette='pastel')
plt.xticks(rotation=45)
plt.title('Migration Distribution by Nation (Violin Plot)')
plt.tight_layout()
plt.show()
```
