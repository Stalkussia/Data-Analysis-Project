```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data.xlsx")
df.head()
```

```python
df = pd.read_excel("data.xlsx",skiprows=3)
df.head()
```

```python
df = df[~df['Items'].isin(['Total arrivals', 'Of living in CIS countries'])]
df.head()
```

```python
df.rename(columns={'Items': 'Country'}, inplace=True)
```

```python
df
```

```python
df = df.loc[:, ['Country'] + list(range(2011, 2025))]
```

```python
df_long = df.melt(id_vars='Country', var_name='Year', value_name='Arrivals')
df_long['Year'] = df_long['Year'].astype(int)
df_long.dropna(inplace=True)
df
```

```python
df_long['Arrivals'].replace("-", pd.NA, inplace=True)
```

```python
df_long['Arrivals'] = pd.to_numeric(df_long['Arrivals'], errors='coerce')
df_long.dropna(subset=['Arrivals'], inplace=True)
df_long
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("Set2")

sns.lineplot(data=df_long, x='Year', y='Arrivals', hue='Country', marker='o')

plt.title('Migration to Kyrgyzstan by Country of Origin (2011–2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
plt.tight_layout()
plt.show()
```

```python
df_area = df_long.pivot(index='Year', columns='Country', values='Arrivals')
df_area.fillna(0, inplace=True)

df_area.plot.area(figsize=(14, 7), cmap='tab20')
plt.title('Migration to Kyrgyzstan by Country (Stacked Area, 2011–2024)')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
heatmap_data = df_long.pivot(index='Country', columns='Year', values='Arrivals')

sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False, linewidths=0.5)
plt.title('Migration to Kyrgyzstan by Country and Year (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
df_long_final = df_long_filtered[
    ~df_long_filtered['Country'].isin(['Other CIS countries'])
]
```

```python
heatmap_data = df_long_final.pivot(index='Country', columns='Year', values='Arrivals')

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)
plt.title('Migration to Kyrgyzstan by Country and Year (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Pivot для area plot
df_area = df_long_final.pivot(index='Year', columns='Country', values='Arrivals')
df_area.fillna(0, inplace=True)  # на случай пропусков

# Строим график
df_area.plot.area(figsize=(14, 7), cmap='tab20')
plt.title('Migration to Kyrgyzstan by Country (Stacked Area, 2011–2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Arrivals')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
plt.tight_layout()
plt.show()

```

```python
year = 2023
top = df_long_final[df_long_final['Year'] == year].sort_values('Arrivals', ascending=False).head(5)

plt.figure(figsize=(8, 5))
sns.barplot(data=top, x='Arrivals', y='Country', palette='viridis')
plt.title(f'Top 5 Countries Sending Migrants to Kyrgyzstan in {year}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_long_final, x='Country', y='Arrivals')
plt.xticks(rotation=45)
plt.title('Distribution of Annual Arrivals by Country (2011–2024)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df_long_final, x='Country', y='Arrivals', jitter=True, size=6, palette='Set1')
plt.xticks(rotation=45)
plt.title('Yearly Migration Points by Country')
plt.tight_layout()
plt.show()
```

```python
df_cum = df_long_final.pivot(index='Year', columns='Country', values='Arrivals').fillna(0).cumsum()

df_cum.plot(figsize=(14, 6))
plt.title('Cumulative Migration to Kyrgyzstan by Country (2011–2024)')
plt.xlabel('Year')
plt.ylabel('Total Arrivals')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
year = 2023
pie_data = df_long_final[df_long_final['Year'] == year].groupby('Country')['Arrivals'].sum()
pie_data = pie_data[pie_data > 0]

plt.figure(figsize=(8, 8))
plt.pie(pie_data, labels=None, startangle=140)
plt.legend(pie_data.index, loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Migration Share by Country in {year}')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_long_final, x='Country', y='Arrivals', inner='quartile', palette='pastel')
plt.xticks(rotation=45)
plt.title('Migration Distribution by Country (Violin Plot)')
plt.tight_layout()
plt.show()
```
